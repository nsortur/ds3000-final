```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('final.csv')
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# only choose entries that are within certain bounds to get a good visual overview
not_outliers = df[df['Adj. Close'] < 100]
print(
    f'{not_outliers.shape[0]} examples in the not-outlier set. {round(not_outliers.shape[0] / df.shape[0], 3)}% of data captured.')
print(f'Mean of not-outlier set: {not_outliers["Adj. Close"].mean()}')
print(f'Standard dev. of not-outlier set: {not_outliers["Adj. Close"].std()}')
fig12 = px.histogram(not_outliers, x='Adj. Close', marginal='box', template='ggplot2',
                     title='Distribution of Adjusted Closing Prices, Without Outliers', nbins=40)
fig12.show()
```

```python
# central tendencies of all the data, including outliers
print(f'Mean of all data: {round(df["Adj. Close"].mean(), 2)}')
print(f'Median of all data: {df["Adj. Close"].median()}')
print(f'Standard dev. of not-outlier set: {round(df["Adj. Close"].std(), 2)}')
```
